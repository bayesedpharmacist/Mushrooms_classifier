### Creating test dataset

```python
import os
import shutil
import numpy as np
```

```python
source_dir = 'original_data/Sample_data'
```

```python
dest_dir ='original_data/Test_data'
```

```python
os.listdir(source_dir)
```

```python
files = os.listdir(source_dir)
```

```python
full_path = []
for root, dirs, files in os.walk(source_dir):
    for d in dirs:
        full_path.append(os.path.join(root, d))    
```

```python
#os.listdir(full_path[5])
full_path
```

```python
for i in full_path:
    files = os.listdir(i)
    for f in files:
        if np.random.rand(1) < 0.1:
            shutil.move(i + '/'+ f, dest_dir + '/'+ f)
```
